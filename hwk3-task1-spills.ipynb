```python
# Import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd

# Import New York oil spill data
fp = os.path.join("data" , "Spill_Incidents_20251107.csv")
spills = pd.read_csv(fp)

# Import counties data
fp = os.path.join("data", "tl_2023_us_county", "tl_2023_us_county.shp")
counties = gpd.read_file(fp)

# Make all column names lower case, remove any extra whitespace and replace spaces with underscores
spills.columns = spills.columns.str.lower().str.strip().str.replace(" ", "_")

# Convert spill data to date time type
spills["spill_date"] = pd.to_datetime(spills["spill_date"])

# Subset data frame by date slice and material
petr_23 = (spills[(spills["spill_date"] >= "2023-01-01") &
                 (spills["spill_date"] <= "2023-10-31") &
                 (spills["material_family"] == "Petroleum")])

# Aggregate petroleum data to summarize the number of spills per county 
spills_per_county = petr_23.groupby("county").size().to_frame(name = "n_spills").reset_index()

# Make all column names lower case and use strip to remove any extra whitespaces if necessary
counties.columns = counties.columns.str.lower().str.strip()

# Selecting only counties from state code 36 (New York)
ny_counties = counties[counties["statefp"] == "36"]

# Replace "St" with "St." for a value in the county column
spills_per_county.county = spills_per_county.county.str.replace("St Lawrence", "St. Lawrence")

# Updating the name of the counties column to match what it is in spills data
ny_counties = ny_counties.rename(columns = {"name":"county"})

# Overwrite ny_counties to include spill data by the county name column
ny_counties = pd.merge(ny_counties, spills_per_county, 
                            how = "inner", 
                            on = "county")

# Add 1 spill count in Niagara county to account for the Tuscarora Indian Reservation
ny_counties.loc[ny_counties["county"] == "Niagara", "n_spills"] += 1

# Plot spill data in New York counties
fig, ax = plt.subplots()
ny_counties.plot(ax=ax, column = "n_spills", # Shade by number of spills in each county
                 legend=True, # Show legend
                 cmap = "rainbow", 
                 edgecolor="grey")
plt.axis(False) # Remove default axis
plt.title("2023 Petroleum Spills in New York by County") # Add title
fig.text(0, 0, "Accessed 2025-11-08 from New York State Department of Environmental Conservation") # Add caption
plt.show()
```
